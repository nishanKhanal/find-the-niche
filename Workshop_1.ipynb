```python
from sklearn.datasets import load_wine
wdf = load_wine(as_frame=True)
wdf = wdf.frame
wdf.head()
```

```python
wdf.info()
```

## Box Plot of Acholor by Wine Type

```python
import matplotlib.pyplot as plt
wdf.boxplot(column='alcohol', by='target')
plt.title('Box Plot of Alcohol by Target')
plt.suptitle('')
plt.xlabel('Target')
plt.ylabel('Alcohol')
plt.show()
```

## Scatter Plot of Malic acid vs Alcohol

```python
wdf.plot(kind='scatter', x='alcohol', y='malic_acid')
plt.title("Malic acid vs alcohol")
plt.show()
```

## Heatmap of the Correlation among all the variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(wdf.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Diabetes Dataset')
plt.show()
```
